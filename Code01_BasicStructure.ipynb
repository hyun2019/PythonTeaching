#### 집합 (set)
- list, tuple 과의 가장 중요한 차이점은 
    1. 중복을 허용하지 않는다는 점
    2. 순서를 보존하지 않는다는 점
- 어떤 변수가 포함되어 있는지를 확인할 때에는 집합이 빠름
- 집합은 { } 로 정의

```python
A_List  = [1,1,1]
A_Tuple = (1,1,1)
A_Set   = {1,1,1}
print(A_List)
print(A_Tuple)
print(A_Set)
```

```python
A_List  = [3,2,1]
A_Tuple = (3,2,1)
A_Set   = {3,2,1}
print(A_List)
print(A_Tuple)
print(A_Set)
```

### <font color='red'>중요한 특징</font>
- 두 변수를 = 으로 정의하고 하나를 변화시키면 다른 하나도 변화
- 두 변수가 메모리상에서 동일한 위치를 차지하기 때문

```python
X = [1]
Y = X
```

```python
print(X)
```

```python
print(Y)
```

```python
Y.append(2)
print(Y)
```

```python
print(X)
```

```python
X = [1]
Y = X.copy()
Y.append(2)
print(X)
print(Y)
```

```python
A0 = A1 = [1,[2],3]
A0[1].append(4)
print(A0)
print(A1)
```

```python
A0 = [1,[2],3]
A1 = A0.copy()
A0[1].append(4)
print(A0)
print(A1)
```

```python
from copy import deepcopy
A0 = [1,[2],3]
A1 = deepcopy(A0)
A0[1].append(4)
print(A0)
print(A1)
```

```python
D0 = D1 = {"A": [1]}
D0["A"].append(4)
print(D0)
print(D1)
```

```python
D0 = D1 = [{"A": [1]}]
D0[0]["B"] = 1
print(D0)
print(D1)
```

### 4. Loop 와 Loop 관련 명령어

- 일정 범위 숫자만큼 loop

```python
for i in range(0,5):
    print(i)
```

- 계속 loop

```python
while True:
    print("1")
```

- list 안의 element 들에 대해 loop
    * tuple, set 역시 기본적 구조는 동일

```python
A = [1,2,3]
for a in A:
    print(a)
```

- dictionary key, value 에 대해 loop

```python
A = {"a":1, "b":2, "c":3}
```

```python
for key, value in A.items():
    print("key:", key, "  value:", value)
```

꼭 key, value 여야할 필요는 없음

```python
for k, v in A.items():
    print("key:", k, "  value:", v)
```
